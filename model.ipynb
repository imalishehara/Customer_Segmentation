```python
data['Returns'] = data['Returns'].fillna(0)
```

```python
data = data.drop(columns=['Customer Name', 'Customer ID', 'Purchase Date', 'Customer Age'])
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
import joblib

# Create separate LabelEncoders
le_category = LabelEncoder()
le_payment = LabelEncoder()
le_gender = LabelEncoder()

# Fit and transform
data['Product Category'] = le_category.fit_transform(data['Product Category'])
data['Payment Method'] = le_payment.fit_transform(data['Payment Method'])
data['Gender'] = le_gender.fit_transform(data['Gender'])

# Save LabelEncoders
joblib.dump(le_category, "le_category.pkl")
joblib.dump(le_payment, "le_payment.pkl")
joblib.dump(le_gender, "le_gender.pkl")
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
features_to_scale = ['Total Purchase Amount', 'Quantity', 'Returns', 'Age', 'Product Price']
data[features_to_scale] = scaler.fit_transform(data[features_to_scale])

# Save scaler
joblib.dump(scaler, "scaler.pkl")
```

```python
from sklearn.cluster import KMeans


X_for_cluster = data[['Product Price','Quantity','Total Purchase Amount','Returns']]
kmeans = KMeans(n_clusters=4, random_state=42)
data['Cluster'] = kmeans.fit_predict(X_for_cluster)


joblib.dump(kmeans, "customer_segment_model.pkl")

cluster_summary = data.groupby('Cluster').mean()
print(cluster_summary[['Product Price','Quantity','Total Purchase Amount','Returns']])
```

```python
from imblearn.over_sampling import SMOTE

X = data.drop(['Churn'], axis=1)
y = data['Churn']

# Balance classes using SMOTE
smote = SMOTE(random_state=42)
X_balanced, y_balanced = smote.fit_resample(X, y)


print(pd.Series(y_balanced).value_counts())

```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_balanced, y_balanced, test_size=0.2, random_state=42
)

```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Save model
joblib.dump(model, "churn_rf_model.pkl")

```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score

y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:,1]

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, y_prob))

```
